```python
# Inisialisasi saldo awal 0
saldo = 0

# Buat list kosong buat nyimpen semua catatan transaksi
catatan = []

# Program utama, dijalankan terus selama belum milih keluar
while True:
    # Nampilin menu pilihan di layar
    print("\n=== Aplikasi Catatan Duit ===")
    print("1. Tambah Pemasukan")
    print("2. Tambah Pengeluaran")
    print("3. Lihat Catatan Transaksi")
    print("4. Cek Saldo")
    print("5. Keluar")

    # User input pilih menu (harus angka 1-5)
    pilih = input("Pilih menu (1-5): ")

    # Kalau user pilih 1, berarti mau nambah duit masuk
    if pilih == "1":
        # Input jumlah duit masuk
        duit = int(input("Masukin duit masuk: Rp "))
        # Input keterangan buat ngasih tahu duit dari mana
        keterangan = input("Keterangan: ")
        # Tambahin duit masuk ke saldo
        saldo += duit
        # Simpan transaksi ke list catatan
        catatan.append(("Masuk", duit, keterangan))

    # Kalau pilih 2, berarti mau ngurangin duit (pengeluaran)
    elif pilih == "2":
        # Input jumlah duit keluar
        duit = int(input("Masukin duit keluar: Rp "))
        # Input keterangan buat ngasih tahu buat apa duitnya keluar
        keterangan = input("Keterangan: ")
        # Kurangin saldo sama duit keluar
        saldo -= duit
        # Simpan transaksi ke list catatan
        catatan.append(("Keluar", duit, keterangan))

    # Kalau pilih 3, liat semua catatan transaksi yang udah dicatet
    elif pilih == "3":
        print("\n=== Catatan Transaksi ===")
        # Tampilkan semua transaksi satu-satu
        for jenis, jumlah, ket in catatan:
            print(f"{jenis}: Rp {jumlah} - {ket}")

    # Kalau pilih 4, cuma pengen liat saldo terakhir
    elif pilih == "4":
        print(f"\nSaldo sekarang: Rp {saldo}")

    # Kalau pilih 5, keluar dari program
    elif pilih == "5":
        print("Sip! Program selesai.")
        break  # Berhenti dari while loop, program selesai

    # Kalau masukin angka selain 1-5, munculin peringatan
    else:
        print("Input salah, coba lagi.")
```
